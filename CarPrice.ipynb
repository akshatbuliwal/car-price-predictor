```python
#cleaning tasks

# remove many non year values from year 
# convert year object to int 
#remove ask for price from Price 
#remove commas from price
# convert price object to integer 
# remove petrol and nan valus from kms_driven 
#remove kms and , from kms_driven 
# remove nan values from fuel_type
#fetch first 3 words from the name of the car
```

```python
car = car[car['year'].astype(str).str.isnumeric()]
car['year'] = car['year'].astype(int)
```

```python
car['year'].info()
```

```python
car['Price'] = car['Price'].astype(str)  
car = car[car['Price'].str.lower() != 'ask for price']
car['Price'] = car['Price'].str.replace(',', '', regex=False)  
car['Price'] = car['Price'].astype(int)  
```

```python
car['Price'].info()
```

```python
car['kms_driven'] = car['kms_driven'].astype(str)
car = car[car['kms_driven'].notna()] 
car = car[~car['kms_driven'].str.lower().str.contains('petrol')]  
car['kms_driven'] = car['kms_driven'].str.replace(' kms', '', regex=False)
car['kms_driven'] = car['kms_driven'].str.replace(',', '', regex=False)
car['kms_driven'] = car['kms_driven'].astype(int)
```

```python
car = car[car['fuel_type'].notna()] 
```

```python
car['name'].unique()
```

```python
car['name'].isna().any()
```

```python
car['name'] = car['name'].str.split().str.slice(0, 3).str.join(' ')
```

```python
car.head(10)
```

```python
car.info()
```

```python
car.describe()
```

```python
# 75% cars priced around 5 lakh 
# maximum price 85 lakhs , so filtering out cars more expensive than 6 lakh as extreme values in regression algorithms can act like outliers 

car = car[car['Price'] < 600000]
```

```python
car.head()
```

```python
car.to_csv('Cleaned_Car_data.csv')
car.info()
# data cleaning ends here 
```

```python
# data preprocessing starts here 
X=car[['name','company','year','kms_driven','fuel_type']]
Y=car['Price']
```

```python
X
```

```python
#categorical columns need to be converted into numerical columns using OneHotEncoder 
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore')
ohe.fit(X[['name', 'company', 'fuel_type']])
X_encoded = ohe.transform(X[['name', 'company', 'fuel_type']]).toarray()
```

```python
# remaining numerical columns need to be scaled to improve model performance 
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
X_remaining = X.drop(columns=['name','company','fuel_type'])
X_scaled = scaler.fit_transform(X_remaining)
```

```python
# now concatenating the encoded and scaled columns to get the final input 
X_final = np.concatenate([X_encoded,X_scaled], axis=1)
```

```python
# Model building
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

X_train,X_test,Y_train,Y_test = train_test_split(X_final,Y,test_size=0.2,random_state = 42)
lr = LinearRegression()
lr.fit(X_train , Y_train) #training the model
Y_pred = lr.predict(X_test)

print("R2 score:", r2_score(Y_test, Y_pred))
```

```python
# Searching for best random_state
scores = []
for i in range(1000):
    X_train, X_test, Y_train, Y_test = train_test_split(X_final, Y, test_size=0.2, random_state=i)
    lr = LinearRegression()
    lr.fit(X_train, Y_train)
    Y_pred = lr.predict(X_test)
    scores.append(r2_score(Y_test, Y_pred))
```

```python
#fetching the best scores and best state 

best_score = max(scores)
best_state = scores.index(best_score)


print("Best R2 score:", best_score)
print("Best random_state:", best_state)
```

```python
# using the best random state for final model building
X_train,X_test,Y_train,Y_test = train_test_split(X_final,Y,test_size=0.2,random_state = best_state)
lr = LinearRegression()
lr.fit(X_train , Y_train) #training the model
Y_pred = lr.predict(X_test)


print("R2 score of the final model / best R2 score:", best_score)
```

```python
import pickle

# Save the trained model
with open('LinearRegressionModel.pkl', 'wb') as f:
    pickle.dump(lr, f)

# Save the OneHotEncoder
with open('OneHotEncoder.pkl', 'wb') as f:
    pickle.dump(ohe, f)

# Save the StandardScaler
with open('StandardScaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
```
